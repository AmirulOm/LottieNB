### Quick render via Line Magic

We can render a lottie using default parameters using
line magic

```python
# Render from file
%render_file ./samples/alert.json

# Render raw content
with open("samples/alert.json", "r") as f:
    stuff = f.read()

%render_json $stuff

# Render from url
%render_url https://assets3.lottiefiles.com/packages/lf20_arq6m8vt.json
```

### Render via methods
Alternatively, for better configurability we can also
render via the methods directly

```python
render_file("samples/alert.json")
render_json(stuff)
render_url("https://assets3.lottiefiles.com/packages/lf20_arq6m8vt.json")
```
